## Downloading Audio files from Deezer collected data

```python
# importing required packaegs
import pandas as pd
import json
import urllib.request as req
from tqdm import tqdm
```

```python
artist_df = pd.read_pickle('Deezer Data/artists.pkl')
artist_df.shape
```

```python
artist_df.head()
```

```python
# downloading songs from a url and saving in mp3 format
artist_mp4_mapping = dict(zip(artist_df['song_id'], artist_df['preview']))
for idx, url in tqdm(artist_mp4_mapping.items()):
    try:
        download_path = 'Deezer Data/artist/mp3/' + str(idx) +'.mp3'
        req.urlretrieve(url, download_path)
    except:
        print('Error ocuured for Id:', idx)
```
